```python
df_expo_acum=pd.read_csv("https://raw.githubusercontent.com/FernandoGuillermo/Training2022-/main/expo_peras_argentinas/vizualizaciones/expo_peras_acum.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
# Dibuja la capa base
fig, ax1 = plt.subplots(figsize=(10,6))
color = 'tab:green'
# Primer gr√°fico de barras del gr√°fico
ax1 = sns.barplot(x='periodo_mens', y='valor_acum', data = df_expo_acum, palette='summer')
ax1.tick_params(axis='y')
 #twinxShare eje x (sintaxis similar, como compartir el eje y twiny)
ax2 = ax1.twinx()
color = 'tab:red'
 # Segundo gr√°fico, gr√°fico de l√≠neas
ax2 = sns.lineplot(x='periodo_mens', y='expo_valor', data = df_expo_acum, sort=False, color=color)
ax2.tick_params(axis='y', color=color)
 #Visualizar resultados de dibujo
plt.show()
```
